```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from io import BytesIO

# Interactive widgets
from ipywidgets import widgets
```

```python
IMG_SIZE = (224, 224)  # Image size (height, width)
BATCH_SIZE = 32  # Batch size
```

```python
# Load the model
#model = tf.keras.models.load_model("../models/mobilenetv3_classifier.keras")
model = tf.keras.models.load_model("../models/densenet_classifier.keras")

# Verify the model structure
model.summary()
```

```python
rescale_layer = tf.keras.layers.Rescaling(scale=1.0 / 255)
```

```python
# Create the widget and take care of the display
uploader = widgets.FileUpload(accept="image/*", multiple=True)
display(uploader)
out = widgets.Output()
display(out)

def file_predict_and_plot(filename, file, out):
    """
    A function for predicting and displaying the image alongside its classification result.
    """
    # Load and preprocess the image
    image = tf.keras.utils.load_img(file, target_size=IMG_SIZE)
    image_array = tf.keras.utils.img_to_array(image)
    image_scaled = rescale_layer(image_array)
    image_scaled = np.expand_dims(image_scaled, axis=0)

    # Get prediction
    prediction = model.predict(image_scaled, verbose=0)[0][0]

    # Display the image and prediction
    with out:
        plt.figure(figsize=(4, 4))
        plt.imshow(image)
        plt.axis("off")
        print (prediction)
        if prediction <= 0.5:
            plt.title(f"{filename}: Chart")
            print(f"{filename} is a chart")
        else:
            plt.title(f"{filename}: Not a Chart")
            print(f"{filename} is not a chart")
        plt.show()


def on_upload_change(change):
    """A function for geting files from the widget and running the prediction."""
    # Get the newly uploaded file(s)

    items = change.new
    for item in items:  # Loop if there is more than one file uploaded
        file_jpgdata = BytesIO(item.content)
        file_predict_and_plot(item.name, file_jpgdata, out)


# Run the interactive widget
# Note: it may take a bit after you select the image to upload and process before you see the output.
uploader.observe(on_upload_change, names="value")
```
